```python
import cv2
```

```python
face=cv2.CascadeClassifier(r"C:\Users\Erik\Downloads\haarcascade_frontalface_alt.xml")
# img = cv2.imread(r"C:\Users\Erik\Pictures\KARUNA.jpg",0)
```

```python
vid=cv2.VideoCapture(0)
face_data= []
count=0
while True:
    suc, img = vid.read()
    print(suc)
    gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    arr=face.detectMultiScale(gray_img)
    for x,y,w,h in arr:
        crop_img=gray_img[y-10:y+h+10, x-10:x+w+10]
        
    cv2.imshow("crop",crop_img)
    cv2.imshow('gray',gray_img)
    

    if count%20==0:
        face_data.append(cv2.resize(crop_img,(100,100)))
        
    if cv2.waitKey(1) == ord('q'):
        break
cv2.destroyAllWindows()
```
