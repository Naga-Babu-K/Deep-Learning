```python
x = []
y = []

for i in range(len(data)):
    # print(i)
    for j in range(len(data[i])-1):
        # print(j, end=' ')
        # print(data[i][j])
        x.append(data[i][:j+1])
        y.append(data[i][j+1])
```

```python
from keras.utils import pad_sequences, to_categorical
```

```python
fv = pad_sequences(x)
```

```python
fv.shape
```

```python
cv = to_categorical(y, num_classes=1712+1)
```

```python
fv
```

```python
cv
```

```python
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN, LSTM, GRU, Bidirectional, Embedding
```

```python
model = Sequential()
model.add(Embedding(1713, 10, input_shape=(114,)))
model.add(LSTM(100, activation='tanh', return_sequences=False))
model.add(Dense(1713, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(fv, cv, batch_size=50, epochs=150)
```
